```python
import numpy as np
```

Part 1: Creating and Reshaping Arrays

```python
arr1 = np.linspace(0,50,10)
arr2 = np.random.randint(5,20,9)
arr2 = arr2.reshape(3,3)
print(arr1)
print(arr2)
```

```python
print(arr1.ndim)
print(arr2.ndim)
print(arr1.shape)
print(arr2.shape)
```

```python
arr1 = arr1.reshape(2,5)
print(arr1)
arr1= arr1.reshape(-1)
print(arr1)
```

Part 2: Indexing and Slicing

```python
#Indexing
print(arr2)
print(arr2[1][2])
print(arr2[0])
print(arr2[1])

#Slicing
print(arr2[0:2, 1:])
print(arr1[0:4])
print(arr1[::-1])
```

Part 3: Array Operations

```python
arr3 = np.arange(1,11)
print(np.add(arr1,arr3))
print(np.multiply(arr1,arr3))
print(np.subtract(arr3,5))
print(arr1+arr3)
print(arr1*arr3)
```

```python
arr4 = np.random.rand(5)
print(arr4)
arr5 = np.random.randint(-1,1,5)
print(arr5)
```

Part 4: Random Number Generation and Copying
