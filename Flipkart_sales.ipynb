```python
import pandas as pd
import numpy as np
```

Loading the dataset

```python
df= pd.read_csv('https://github.com/Taj1920/Data-Analysis-Datasets/raw/refs/heads/main/Datasets-Practice/flipkart_sales.csv')
df
```

```python
#inbuilt methods of df
df.head()
df.info()
df.describe()
df.shape
# df['Location'].unique() # This line caused the error

# To inspect available columns, you can use:
df.columns
# Or, if you meant to get unique values from a different column, replace 'Location' with an existing column name, for example:
df['Category'].unique()
```

```python
#Applying boolean filtering on dataframes
#syntax: df[condition]
#1. get data of books cat
df[df['Category']=='Books']
#2. get the records where price > 5000
df[df['Price (INR)']>5000]
#3. smart watch
df[df['Product Name']=='Smartwatch']
#4. prod name where price > 30000 and cust rating atleast 3.0
df.loc[(df['Price (INR)'] > 30000) & (df['Customer Rating'] >= 3.0), 'Product Name']
#5. fetch data of Face cream where price < 20000 and quantity >3
df[(df['Product Name'] == 'Face cream') &
   (df['Price (INR)'] < 20000) &
   (df['Quantity Sold'] > 3)]
```

```python
#Advanaced slicing on df
#1.loc: to slice through multiple cols at the same time
#suntax: df.loc[row_slice,[col1,col2......]]
#df.loc[0:10,['Product Name','Price (INR)']]

#2.iloc: also known as index based slicing
#syntax: df.iloc[row_slice,col_slice]
#note: in case of iloc ending index is exclusive
df.iloc[0:11,0:2]
#1. get data from first 100 rows and even indexed colums
df.iloc[0:100,::2]

#2.get data from product name, price, order date from odd indexed rows
df.iloc[1::2,[1,3,6]]
#3.rev all rowss and cols
df.iloc[::-1,::-1]
```

```python
#Group by: if we want to perform some op based on a category we an use group by
#syntax:
#var = df.groupby('cat_col') -> groups the data
#var.groups -> returns a dict with grouped index
#var1=var.get_group("group_name") -> returns specified group
#categorical col: It consists of less no. of unique values

cat = df.groupby('Category')
cat.groups
beauty = cat.get_group('Beauty')
beauty['Category'].count()
#1.find avg price of clothing cat
clothing = cat.get_group('Clothing')
clothing['Price (INR)'].mean()
#2.find total sales of electronics cat
electronics = cat.get_group('Electronics')
electronics['Total Sales (INR)'].sum()
#3.find total sales for all cat
cat['Total Sales (INR)'].sum()
#4.Find avg price for all prod
df.groupby('Product Name')['Price (INR)'].mean()
#5.find avg rating for all cat
cat['Customer Rating'].mean()
#6.find total sales for all paument methods
df.groupby('Payment Method')['Total Sales (INR)'].sum()
#7.which payment type has max total sales
df.groupby('Payment Method')['Total Sales (INR)'].sum().sort_values(ascending=False).head(1)
#or
df.groupby('Payment Method')['Total Sales (INR)'].sum().idxmax()
#or
pay = df.groupby('Payment Method')['Total Sales (INR)'].sum()
pay = pd.DataFrame(pay)
pay.reset_index(inplace=True)
pay.max()
#8.which product got highest sales
df.groupby('Product Name')['Total Sales (INR)'].sum().sort_values(ascending=False).head(1)

#9.which product got lowest sales
df.groupby('Product Name')['Total Sales (INR)'].sum().sort_values(ascending=True).head(1)
#10.which  prod got sold more
df.groupby('Product Name')['Quantity Sold'].sum().sort_values(ascending=False).head(1)
df.groupby('Product Name')['Quantity Sold'].sum().idxmax()
```
